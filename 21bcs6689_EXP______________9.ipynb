```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
text = """
Natural language processing (NLP) is a field of artificial intelligence that focuses on the interaction between computers and humans through natural language. It encompasses a variety of tasks, including text summarization, sentiment analysis, and machine translation.
"""
```

```python
# Get unique characters from the text
chars = sorted(list(set(text)))
char_to_idx = {char: idx for idx, char in enumerate(chars)}
idx_to_char = {idx: char for char, idx in char_to_idx.items()}

# Define maximum sequence length
max_sequence_length = 100

# Generate sequences and next characters
sequences = []
next_chars = []
for i in range(len(text) - max_sequence_length):
    sequences.append(text[i:i + max_sequence_length])
    next_chars.append(text[i + max_sequence_length])

# Create one-hot encoded input sequences and next characters
X = np.zeros((len(sequences), max_sequence_length, len(chars)), dtype=bool)
y = np.zeros((len(sequences), len(chars)), dtype=bool)
for i, sequence in enumerate(sequences):
    for t, char in enumerate(sequence):
        X[i, t, char_to_idx[char]] = 1
    y[i, char_to_idx[next_chars[i]]] = 1
```

```python
model = Sequential([
    LSTM(128, input_shape=(max_sequence_length, len(chars))),
    Dropout(0.2),
    Dense(len(chars), activation='softmax')
])
model.compile(optimizer='adam', loss='categorical_crossentropy')
```

```python
model = Sequential([
    LSTM(128, input_shape=(max_sequence_length, len(chars))),
    Dropout(0.2),
    Dense(len(chars), activation='softmax')
])
model.compile(optimizer='adam', loss='categorical_crossentropy')
model.fit(X, y, batch_size=128, epochs=100)
```

```python
# Define seed text
seed_text = "Natural language processing (NLP) is"
generated_text = seed_text

# Generate text character by character
for _ in range(100):
    X_pred = np.zeros((1, max_sequence_length, len(chars)))
    for t, char in enumerate(seed_text):
        X_pred[0, t, char_to_idx[char]] = 1
    prediction = model.predict(X_pred, verbose=0)[0]
    next_index = np.random.choice(len(chars), p=prediction)
    next_char = idx_to_char[next_index]
    generated_text += next_char
    seed_text = seed_text[1:] + next_char

print("Generated Text:")
print(generated_text)
```
